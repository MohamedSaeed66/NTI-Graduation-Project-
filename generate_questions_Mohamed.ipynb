```python
from openai import OpenAI
import os

os.environ["OPENROUTER_API_KEY"] = "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"

client = OpenAI(
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1"
)
```

```python
def build_prompt(matched_keywords):
    keywords_text = ", ".join(matched_keywords)

    prompt = f"""
You are a professional technical interviewer.

The following are matched keywords extracted from a candidate's CV and the job requirements:
{keywords_text}

Task:
Generate five clear, concise, and professional technical interview questions that evaluates the candidate's understanding of these keywords.

Rules:
- Generate  five question
- No explanations
- No bullet points
- Output must be a single question sentence
-ask questions related to skilles not about projects

Interview Question:
"""
    return prompt
```

```python
def generate_interview_question(matched_keywords):
    prompt = build_prompt(matched_keywords)

    response = client.chat.completions.create(
        model="openai/gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are an AI interviewer assistant."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.7,
        max_tokens=100
    )

    return response.choices[0].message.content.strip()
```

```python
if __name__ == "__main__":

    matched_keywords = [
        "Python",
        "Machine Learning",
        "Data Science",
        "Web Development",
        "Data Visualization",
        "Data Analysis",
        "Power BI"
    ]

    question = generate_interview_question(matched_keywords)

    print("Generated Interview Question:")
    print(question)
```
